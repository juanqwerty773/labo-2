```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("tercera clase (05_09)/csv's/tercera clase (05_09) - Hoja 1.csv")
df.head()
```

```python
distancia = pd.to_numeric(df["distancia E-R (cm)"], errors='coerce')
df["Vpp_rec (mV)"] = df["Vpp_rec (mV)"].str.replace(',','.', regex=False)
df["errores Vpp (mV)"] = df["errores Vpp (mV)"].str.replace(',','.', regex=False)
Vpp_rec = pd.to_numeric(df["Vpp_rec (mV)"], errors='coerce')
error_Vpp = pd.to_numeric(df["errores Vpp (mV)"], errors='coerce')

x = np.linspace(5,50,100)

plt.semilogy(x, 1072/((x-4)**(1/2)), color='blue')
plt.errorbar(distancia, Vpp_rec, yerr=error_Vpp,ecolor = 'black',capsize= 4)
plt.scatter(distancia, Vpp_rec, label='Vpp (mV)',s=10,color='r')
plt.ylim(50,1200)
plt.xlabel('Distancia (cm)')
plt.ylabel('Vpp (mV)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df = pd.read_csv("tercera clase (05_09)/csv's/tercera clase (05_09) - Hoja 2.csv")
df.head()
```

```python
distancia = pd.to_numeric(df["distancia E-R (mm)"], errors='coerce')
df["Vpp_rec (V)"] = df["Vpp_rec (V)"].str.replace(',','.', regex=False)
df["error_Vpp"] = df["error_Vpp"].str.replace(',','.', regex=False)
Vpp_rec = pd.to_numeric(df["Vpp_rec (V)"], errors='coerce')
error_Vpp = pd.to_numeric(df["error_Vpp"], errors='coerce')

plt.errorbar(distancia, Vpp_rec, yerr=error_Vpp,ecolor = 'black',capsize= 4)
plt.scatter(distancia, Vpp_rec, label='Vpp (mV)',s=10,color='r')
plt.xlabel('Distancia (cm)')
plt.ylabel('Vpp (mV)')
plt.grid(True)
plt.legend()
plt.show()
```
